```python
sql = """
    SELECT pa.pais, ci.ciudad
    FROM ciudades ci INNER JOIN paises pa ON (ci.pais_id = pa.pais_id)
    ORDER BY pa.pais ASC;
"""
# Los métodos execute() ejecutan la consulta SQL y devuelven el resultado.

cursor.execute( sql )

for registro in cursor:
    print( registro )
```

## pandas

```python
import pandas as pd

df = pd.read_sql(sql, conexion)
df
```

```python
sql = """
    SELECT cl.cliente_id, cl.nombre, cl.apellido, SUM(pa.cantidad) as importe
    FROM pagos as pa INNER JOIN clientes as cl ON (pa.cliente_id = cl.cliente_id)
    GROUP BY cl.cliente_id, cl.nombre, cl.apellido
    HAVING SUM(pa.cantidad) > 100
    ORDER BY importe DESC;
"""

df = pd.read_sql(sql, conexion)
df
```

```python
# utilizar los métodos cursor.close() y connection.close() para cerrar las conexiones 
# PostgreSQL una vez finalizado el trabajo
cursor.close()
conexion.close()
```
